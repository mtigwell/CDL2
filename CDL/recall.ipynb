```python
import csv

def log_results(ep, loss, recall, time):
    with open('errors.csv', mode='w') as f:
        fieldnames = ['Epoch', 'Loss', 'Recall', 'Time']
        writer = csv.DictWriter(f, fieldnames=fieldnames)
        writer.writeheader()
        writer.writerow({'Epoch': ep, 'Loss': loss, 'Recall': recall, 'Time': time})

log_results(1, 0.5, 0.2, 10)
```
