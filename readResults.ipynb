```python
import os
import pandas as pd
```

```python
root_dir = 'C:/Users/mtigw/Desktop/CDL2/CDL/results'
max = 0

for subdir, dirs, files in os.walk(root_dir):
    if not len(files): continue

    var_path = os.path.join(subdir, files[0]) # cdl.log file
    errors_path = os.path.join(subdir, files[1]) #errors.csv

    f = open(var_path, 'r')
    # print(f.read())

    df = pd.read_csv(errors_path)
    max_recall = df.Recall.max()

    if max_recall > max:
        max = max_recall
        print('new maximum of ', max_recall)
        print(f.read())

```

```python
root_dir = 'C:/Users/mtigw/Desktop/CDL2/CDL/results'
min = 1000000000000000000000

for subdir, dirs, files in os.walk(root_dir):
    if not len(files): continue

    var_path = os.path.join(subdir, files[0]) # cdl.log file
    errors_path = os.path.join(subdir, files[1]) #errors.csv

    f = open(var_path, 'r')
    # print(f.read())

    df = pd.read_csv(errors_path)
    min_loss = df.Loss.min()

    if min_loss < min:
        min = min_loss
        print('new min loss of {} at path {}'.format(min_loss, var_path))
        print(f.read())
```
