```python
import os
import pandas as pd
import numpy as np
```

```python
def get_max(root_dir):
    max = 0

    for subdir, dirs, files in os.walk(root_dir):
        if not len(files): continue

        var_path = os.path.join(subdir, files[0]) # cdl.log file
        errors_path = os.path.join(subdir, files[1]) #errors.csv

        f = open(var_path, 'r')
        # print(f.read())

        df = pd.read_csv(errors_path)
        max_recall = df.Recall.max()

        if isinstance(max_recall, np.float64):
            if max_recall > max:
                max = max_recall
                print('new maximum of ', max_recall)
                # print(f.read())
                print(errors_path)
```
