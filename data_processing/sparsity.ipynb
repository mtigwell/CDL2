```python
# remove the top 10% most dense nodes
def create_dataset():
    user_conn = []
    datafile = ''
    filename = 'newusers.dat'

    # get connection threshold
    with open(r"test.dat") as rating_file:
        lines = rating_file.readlines()
        for _, line in enumerate(lines):
            items = line.strip().split(" ")
            user_conn.append(len(items))

    cutoff = int(len(user_conn) * 0.9)
    user_conn.sort()
    thres = user_conn[cutoff] # the threshold for max number of connections for a user to retain in the dataset

    # remove dense users
    with open(r"test.dat") as rating_file:
        lines = rating_file.readlines()
        for _, line in enumerate(lines):
            items = line.strip().split(" ")
            if len(items) > thres:
                continue
            else:
                datafile += line

    # write to new data file
    newfile = open(filename, 'w')
    newfile.write(datafile)
    newfile.close()
```

```python
def calculate_sparsity():
    count = 0
    
    with open(r"../data/citeulike/citeulike-a/mult.dat") as item_info_file:
        item_size = len(item_info_file.readlines())

    with open(r"newfile.dat") as rating_file:
        lines = rating_file.readlines()
        users = len(lines)
        
        for _, line in enumerate(lines):
            items = line.strip().split(" ")
            count += len(items)

    sparsity = (1 - (count / (item_size * users))) * 100

    print(item_size)
    print(count)
    print(users)
    print(sparsity)
```
